Assumes that the transforms package has been installaed in the venv and all manipulations required for cargo and rep_removal were done in the vnev

```python
from huggingface_hub import hf_hub_download
import pyarrow.parquet as pq
import pandas as pd
import os
```

```python
%%time
REPO_ID = "HuggingFaceFW/fineweb"
FILENAME = "data/CC-MAIN-2013-20/000_00000.parquet"
file1=hf_hub_download(repo_id=REPO_ID, filename=FILENAME, repo_type="dataset")
```

```python
#os.path.dirname(file1)
```

```python
%%time
import pyarrow.parquet as pq
import pandas as pd
table = pq.read_table(file1)
table.to_pandas()
```

```python
%%time
from dpk_doc_id.transform_python import DocID
DocID(input_folder= os.path.dirname(file1),
        output_folder= "files-doc-id",
        doc_id_doc_column= "text",
        doc_id_hash_column= "document_id",
        doc_id_int_column= "int_id_column",
        doc_id_start_id= 5).transform()
```
